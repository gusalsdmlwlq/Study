```python
import xml.etree.ElementTree as ET
import time
```

```python
tree = ET.parse("data/kowiki-20200901-pages-articles.xml")
```

```python
root = tree.getroot()
```

```python
t = root.iter('{http://www.mediawiki.org/xml/export-0.10/}page')
```

```python
data = []
```

```python
count = 0
for page in t:
    entry = {}
    for child in page.iter():
        key = child.tag.split("}")[1]
        if key == "page":
            continue
        value = child.text
        entry[key] = value
    data.append(entry)
    count += 1
    if count == 100:
        break
```
