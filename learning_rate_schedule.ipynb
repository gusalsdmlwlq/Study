```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
lrs = []
warmup = 4000
total = 20000
lr = 1e-3
for step in range(1,total):
    lrs.append(np.where(step<warmup, lr*step/warmup, 0.5 * (1+np.cos((step-warmup)*np.pi/(total-warmup))) * lr))
```

```python
lrs[-1]
```

```python
plt.plot(lrs)
```

```python
lrs = []
warmup = 4000
total = 20000
lr = 1e-3
for step in range(1,total):
    lrs.append(lr * warmup ** 0.5 * min(step * warmup ** -1.5, step ** -0.5))
```

```python
plt.plot(lrs)
```
