````python
from matplotlib import pyplot as plt
import numpy as np
````

````python
points = np.random.random(100)

fig = plt.figure(figsize=(5,4))
plt.title("One-dimension dataset", size=15)
plt.xlabel("X label", size=12, labelpad=15)
plt.ylabel("Y label", size=12, labelpad=15)
plt.xlim(0, 80)
plt.ylim(-5, 5)
plt.xticks(np.arange(0, 81, 5), size=12)
plt.yticks(np.arange(-5, 5, 1), size=12)
plt.grid()
plt.plot(points)
plt.show()
````

````python
points1 = np.random.random(100)
points2 = np.random.random(100) * -1

fig = plt.figure(figsize=(5,4))
plt.title("2 one-dimension datasets in one graph", size=15)
plt.xlabel("X label", size=12, labelpad=15)
plt.ylabel("Y label", size=12, labelpad=15)
plt.xlim(0, 80)
plt.ylim(-5, 5)
plt.xticks(np.arange(0, 81, 5), size=12)
plt.yticks(np.arange(-5, 5, 1), size=12)
plt.grid()
plt.plot(points1, label="Dataset 1", color="r")
plt.plot(points2, label="Dataset 2", color="b")
plt.legend(loc="upper right", fontsize=15)
plt.show()
````

````python
points1 = np.random.random(100)
points2 = np.random.random(100) * -1

fig = plt.figure(figsize=(5,4))

plt.subplot(2,1,1)
plt.title("2 one-dimension datasets in each graph", size=15, pad=15)
plt.xlabel("X label", size=12, labelpad=15)
plt.ylabel("Y label", size=12, labelpad=15)
plt.xlim(0, 80)
plt.ylim(-5, 5)
plt.grid()
plt.plot(points1)

plt.subplot(2,1,2)
plt.xlabel("X label", size=12, labelpad=15)
plt.ylabel("Y label", size=12, labelpad=15)
plt.xlim(0, 80)
plt.ylim(-5, 5)
plt.grid()
plt.plot(points2)

# plt.legend(loc="upper right", fontsize=15)
plt.show()
````

````python
xs = np.array([1,2,3,5,8,13,21,34,55])
ys = np.random.random(9) * 10

plt.rcParams["font.family"] = 'NanumGothic'

fig = plt.figure(figsize=(5,4))
plt.title("Two-dimension dataset", size=15)
plt.xlabel("X ", size=12, labelpad=15)
plt.ylabel("Y ", size=12, labelpad=15)
plt.xlim(0, 60)
plt.ylim(0, 15)
plt.xticks(np.arange(0, 61, 5), size=12)
plt.yticks(np.arange(0, 15, 1), size=12)
plt.grid()
plt.plot(xs, ys)
plt.show()

plt.rcParams["font.family"] = 'sans-serif'
````

````python
nums = [100, 120, 80]

fig = plt.figure(figsize=(5,4))
plt.title("Comparison of 3 classes", size=15)
plt.xlabel("Classes", size=12, labelpad=15)
plt.ylabel("Number", size=12, labelpad=15)
plt.ylim(0, 150)
plt.yticks(np.arange(0, 151, 20), size=12)
plt.bar([0, 1, 2], nums, width=0.5, tick_label=["Class 1", "Class 2", "Class 3"], )
plt.show()

fig = plt.figure(figsize=(5,4))
plt.title("Comparison of 3 classes", size=15)
plt.xlabel("Classes", size=12, labelpad=15)
plt.ylabel("Number", size=12, labelpad=15)
plt.ylim(0, 150)
plt.xticks([])
plt.yticks(np.arange(0, 151, 20), size=12)
plt.bar([0], nums[0], width=0.5, label="Class 1")
plt.bar([1], nums[1], width=0.5, label="Class 2")
plt.bar([2], nums[2], width=0.5, label="Class 3")
plt.legend(loc="upper right")
plt.show()

fig = plt.figure(figsize=(10,8))
explode = [0, 0.05, 0.1]
plt.title("Comparison of 3 classes", size=15)
plt.pie(nums, labels=["Class 1", "Class 2", "Class 3"], autopct='%.3f', radius=0.8, explode=explode)
plt.legend(loc="upper left", )
plt.show()
````

````python
points = np.random.normal(size=1000)

fig = plt.figure(figsize=(5,4))
plt.title("Normal distribution", size=15)
plt.xlabel("Value", size=12, labelpad=15)
plt.ylabel("Number", size=12, labelpad=15)
plt.hist(points, bins=20)
plt.show()
````
