```python
import tensorflow as tf
import random
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.random.normal(size=(1000))
X = tf.constant(x,dtype=tf.float32)
y = X + np.random.normal(size=(1000))
W = tf.Variable(np.random.random())
b = tf.Variable(np.random.random())
pred = W*X + b
loss = tf.reduce_sum((pred-y)**2)
train = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)
```

```python
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    for i in range(1000):
        _,loss_ = sess.run([train,loss])
        if i%100==0:
            print(loss_)
    plt.figure(figsize=(5,4))
    plt.plot(X.eval(),y.eval(),"bo")
    plt.plot(np.arange(-5,5),np.arange(-5,5)*W.eval()+b.eval())
    print(W.eval(),b.eval())
```
